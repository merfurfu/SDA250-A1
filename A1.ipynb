## 6 
A stemmed version of the longest sentence (extracted above in 3).


```python
from nltk import word_tokenize
```

```python
longStemSent = max(treatySent, key=len)
```

```python
print(len(longStemSent))
```

```python
def stem(word):
    regexp = r'^(.*?)(ing|ly|ed|ious|ies|ive|es|s|ment)?$'
    stem, suffix = re.findall(regexp, str(word))[0]
    return stem

raw = str(longStemSent)

tokens = word_tokenize(raw)
[stem(t) for t in tokens]
```
